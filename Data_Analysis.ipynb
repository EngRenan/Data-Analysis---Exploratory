```python
df = df.rename(columns={'Annual_Income_(k$)': 'Annual Income', 'Spending_Score': 'Spendig Score'})
df.head()
```

# Começando a explorar os dados graficamente

```python
#Criando um pairplot
sns.pairplot(df.drop('CustomerID', axis=1))
```

```python
#Analisando quantidade por genero
sns.countplot(x='Genre', data=df)
plt.title('Genre Score Distribution')
plt.show()
```

```python
# Analisando idade
sns.histplot(x='Age', data=df, color='purple', bins=20)
plt.title('AGE Distribution')
plt.show()
```

```python
# Analisando por Income
sns.kdeplot(x='Annual Income', data=df, color='green', fill=True)
plt.title('Income Distribution')
plt.show()
```

```python
# spending score por genero
sns.violinplot(x='Genre', y='Spendig Score', data=df)
plt.title('Spending Score by Genero')
plt.show()
```

```python
 # Score por idade
sns.lineplot(x='Age', y='Spendig Score', color="blue", data=df)
plt.title('Score por Idade')
plt.show()
```

```python
# Analisando Score por idade e genero
sns.lineplot(x='Age', y='Spendig Score', hue='Genre', data=df)
plt.title('Score por idade e genero')
plt.show()
```

```python
# Analisando Income por idade e genero
sns.lineplot(x='Age', y='Annual Income', hue='Genre', data=df)
plt.title('Income por idade e genero')
plt.show()
```

```python
# Score por income
plt.hexbin(x='Annual Income', y='Spendig Score', data=df,
gridsize=20, cmap='Blues')
plt.xlabel('Annual Income')
plt.xticks(rotation=45)
plt.ylabel('Spendig Score')
plt.title('Spendig Score')
plt.colorbar()
plt.show()
```
